# 3. Calculate Angles

```python
def calculate_angle(a,b,c):
    a = np.array(a) # First
    b = np.array(b) # Mid
    c = np.array(c) # End
    
    radians = np.arctan2(c[1]-b[1], c[0]-b[0]) - np.arctan2(a[1]-b[1], a[0]-b[0])
    angle = np.abs(radians*180.0/np.pi)
    
    if angle >180.0:
        angle = 360-angle
        
    return angle 
```

```python
import cv2
import mediapipe as mp
import numpy as np

# Function to calculate angle between three points
def calculate_angle(a, b, c):
    a = np.array(a)  # First point
    b = np.array(b)  # Middle point
    c = np.array(c)  # Last point
    
    radians = np.arctan2(c[1] - b[1], c[0] - b[0]) - np.arctan2(a[1] - b[1], a[0] - b[0])
    angle = np.abs(radians * 180.0 / np.pi)
    
    if angle > 180.0:
        angle = 360 - angle
        
    return angle

# Initialize Mediapipe Pose and Drawing utilities
mp_pose = mp.solutions.pose
mp_drawing = mp.solutions.drawing_utils

# Initialize video capture
cap = cv2.VideoCapture(0)

# Curl counter variables
right_counter = 0
left_counter = 0
right_stage = None
left_stage = None

# Mediapipe Pose model
with mp_pose.Pose(min_detection_confidence=0.5, min_tracking_confidence=0.5) as pose:
    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break

        # Recolor image to RGB
        image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        image.flags.writeable = False

        # Make detection
        results = pose.process(image)

        # Recolor back to BGR
        image.flags.writeable = True
        image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)

        # Extract landmarks
        try:
            landmarks = results.pose_landmarks.landmark

            # Right side coordinates
            rshoulder = [landmarks[mp_pose.PoseLandmark.RIGHT_SHOULDER.value].x,
                         landmarks[mp_pose.PoseLandmark.RIGHT_SHOULDER.value].y]
            relbow = [landmarks[mp_pose.PoseLandmark.RIGHT_ELBOW.value].x,
                      landmarks[mp_pose.PoseLandmark.RIGHT_ELBOW.value].y]
            rwrist = [landmarks[mp_pose.PoseLandmark.RIGHT_WRIST.value].x,
                      landmarks[mp_pose.PoseLandmark.RIGHT_WRIST.value].y]

            # Left side coordinates
            lshoulder = [landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER.value].x,
                         landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER.value].y]
            lelbow = [landmarks[mp_pose.PoseLandmark.LEFT_ELBOW.value].x,
                      landmarks[mp_pose.PoseLandmark.LEFT_ELBOW.value].y]
            lwrist = [landmarks[mp_pose.PoseLandmark.LEFT_WRIST.value].x,
                      landmarks[mp_pose.PoseLandmark.LEFT_WRIST.value].y]

            # Calculate angles
            right_angle = calculate_angle(rshoulder, relbow, rwrist)
            left_angle = calculate_angle(lshoulder, lelbow, lwrist)

            # Right arm curl logic
            if right_angle > 160:
                right_stage = "down"
            if right_angle < 30 and right_stage == "down":
                right_stage = "up"
                right_counter += 1

            # Left arm curl logic
            if left_angle > 160:
                left_stage = "down"
            if left_angle < 30 and left_stage == "down":
                left_stage = "up"
                left_counter += 1

            # Visualize angles
            cv2.putText(image, str(int(right_angle)),
                        tuple(np.multiply(relbow, [640, 480]).astype(int)),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2, cv2.LINE_AA)
            cv2.putText(image, str(int(left_angle)),
                        tuple(np.multiply(lelbow, [640, 480]).astype(int)),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2, cv2.LINE_AA)

        except Exception as e:
            print(f"Error: {e}")

        # Render the counter display
        cv2.rectangle(image, (0, 0), (640, 100), (245, 117, 16), -1)

        # Display right reps
        cv2.putText(image, 'RIGHT REPS', (15, 30),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255, 255, 255), 2, cv2.LINE_AA)
        cv2.putText(image, str(right_counter), (15, 70),
                    cv2.FONT_HERSHEY_SIMPLEX, 2, (255, 255, 255), 2, cv2.LINE_AA)

        # Display left reps
        cv2.putText(image, 'LEFT REPS', (330, 30),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255, 255, 255), 2, cv2.LINE_AA)
        cv2.putText(image, str(left_counter), (330, 70),
                    cv2.FONT_HERSHEY_SIMPLEX, 2, (255, 255, 255), 2, cv2.LINE_AA)

        # Render pose landmarks
        mp_drawing.draw_landmarks(image, results.pose_landmarks, mp_pose.POSE_CONNECTIONS,
                                  mp_drawing.DrawingSpec(color=(245, 117, 66), thickness=2, circle_radius=2),
                                  mp_drawing.DrawingSpec(color=(245, 66, 230), thickness=2, circle_radius=2))

        # Display the frame
        cv2.imshow('Mediapipe Feed', image)

        # Exit on 'q'
        if cv2.waitKey(10) & 0xFF == ord('q'):
            break

    cap.release()
    cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft
from scipy.signal import butter, filtfilt
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
import time

# Bandpass filter
def butter_bandpass(lowcut, highcut, fs, order=5):
    nyquist = 0.5 * fs
    low = lowcut / nyquist
    high = highcut / nyquist
    b, a = butter(order, [low, high], btype='band')
    return b, a

def bandpass_filter(data, lowcut, highcut, fs):
    b, a = butter_bandpass(lowcut, highcut, fs)
    return filtfilt(b, a, data)

# Signal processing for rPPG (heart rate calculation from color signal)
def calculate_heart_rate(color_signal, fps):
    signal_fft = fft(color_signal)
    frequencies = np.fft.fftfreq(len(color_signal), d=1/fps)  # Frequency in Hz
    peak_freq = np.abs(frequencies[np.argmax(np.abs(signal_fft))])  # Find the peak frequency
    heart_rate_bpm = peak_freq * 60  # Convert from Hz to BPM
    return heart_rate_bpm

# Fatigue detection logic (using heart rate and HRV)
def calculate_hrv(rr_intervals):
    # Calculate HRV using the standard deviation of RR intervals
    hrv = np.std(rr_intervals)
    return hrv

def detect_peaks(filtered_signal):
    # Simple peak detection based on thresholding and local maxima
    threshold = np.mean(filtered_signal) + 0.5 * np.std(filtered_signal)
    peaks = []
    for i in range(1, len(filtered_signal) - 1):
        if filtered_signal[i] > threshold and filtered_signal[i] > filtered_signal[i-1] and filtered_signal[i] > filtered_signal[i+1]:
            peaks.append(i)
    return peaks

class GymTrainer:
    def __init__(self):
        self.bpm_threshold = 160  # Threshold for high heart rate indicating fatigue
        self.resting_bpm = 70  # Resting BPM for the user (baseline heart rate)
        self.model = self.train_fatigue_detection_model()
    
    def train_fatigue_detection_model(self):
        # Example of training a Random Forest model for fatigue prediction
        X = np.array([[120, 50, 70], [140, 60, 80], [160, 70, 90]])  # [Heart Rate, HRV, Workout Intensity]
        y = np.array([0, 0, 1])  # 0 = Not fatigued, 1 = Fatigued
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33)
        model = RandomForestClassifier()
        model.fit(X_train, y_train)
        return model
    
    def detect_fatigue(self, heart_rate_bpm, hrv, workout_intensity):
        # Predict fatigue using the trained model
        features = np.array([heart_rate_bpm, hrv, workout_intensity]).reshape(1, -1)
        prediction = self.model.predict(features)
        if prediction == 1:
            return True  # Fatigued
        return False  # Not fatigued
    
    def give_advice(self, fatigued):
        if fatigued:
            print("AI Trainer: Take a rest or hydrate. Stretch your muscles.")
        else:
            print("AI Trainer: Keep going! You're doing great!")
    
    def process_heart_rate(self, color_signal, fps, workout_intensity):
        filtered_signal = bandpass_filter(color_signal, lowcut=0.7, highcut=2.0, fs=fps)
        heart_rate_bpm = calculate_heart_rate(filtered_signal, fps)
        print(f"Current Heart Rate: {heart_rate_bpm:.2f} BPM")

        # Detect peaks in the rPPG signal for HRV calculation
        peaks = detect_peaks(filtered_signal)
        rr_intervals = np.diff(peaks)  # RR intervals (differences between consecutive peaks)
        hrv = calculate_hrv(rr_intervals)

        fatigued = self.detect_fatigue(heart_rate_bpm, hrv, workout_intensity)
        self.give_advice(fatigued)

# Initialize the AI-based Gym Trainer
trainer = GymTrainer()

# Initialize video capture (default webcam)
cap = cv2.VideoCapture(0)

# Load pre-trained face detector (Haar Cascade classifier)
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

color_signal = []  # List to store color signal data
fps = 30  # Assuming 30 frames per second
workout_intensity = 1  # Placeholder for workout intensity (can be calculated based on sets/reps)

while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Convert frame to grayscale for face detection
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Detect faces in the frame
    faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))

    for (x, y, w, h) in faces:
        roi = frame[y:y + h, x:x + w]  # Region of interest: the detected face
        avg_color = np.mean(roi, axis=(0, 1))  # Mean of RGB channels
        color_signal.append(avg_color[1])  # Use green channel (for pulse signal)
        
        # Draw rectangle around the detected face
        cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 2)

    # After enough data is collected, process the heart rate and give feedback
    if len(color_signal) > 100:
        trainer.process_heart_rate(color_signal, fps, workout_intensity)
        color_signal = []  # Reset for the next period of analysis

    # Display the frame with face detection rectangle
    cv2.imshow('Frame', frame)

    # Exit loop when 'q' key is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the video capture and close all OpenCV windows
cap.release()
cv2.destroyAllWindows()
```

```python
# updatedddd pullups
import time
import cv2
import numpy as np
import mediapipe as mp

# Initialize MediaPipe pose detector
mp_pose = mp.solutions.pose
mp_drawing = mp.solutions.drawing_utils
pose = mp_pose.Pose(static_image_mode=False, min_detection_confidence=0.5, min_tracking_confidence=0.5)

# Helper functions
def calculate_angle(a, b, c):
    radians = np.arctan2(c[1] - b[1], c[0] - b[0]) - np.arctan2(a[1] - b[1], a[0] - b[0])
    return np.abs(np.degrees(radians))

def is_near_floor(wrist, elbow, threshold=0.1):
    return abs(wrist[1] - elbow[1]) < threshold

def is_user_moving(current_landmarks, previous_landmarks):
    movement_threshold = 0.05
    for i in range(len(current_landmarks)):
        if abs(current_landmarks[i].x - previous_landmarks[i].x) > movement_threshold or \
           abs(current_landmarks[i].y - previous_landmarks[i].y) > movement_threshold:
            return True
    return False

# Initialize workout variables
pushup_counter, pullup_counter, squat_counter = 0, 0, 0
pushup_stage, pullup_stage, squat_stage = None, None, None
start_time, end_time = None, None
last_movement_time = time.time()
rest_times = []
workout_active = False
previous_landmarks = None

# Start video capture
cap = cv2.VideoCapture(0)
cap.set(3, 640)  # Width
cap.set(4, 480)  # Height

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    frame = cv2.flip(frame, 1)
    rgb_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    results = pose.process(rgb_frame)

    if results.pose_landmarks:
        landmarks = results.pose_landmarks.landmark

        # Detect movement
        if previous_landmarks:
            if is_user_moving(landmarks, previous_landmarks):
                if not workout_active:
                    start_time = time.time()
                    workout_active = True
                last_movement_time = time.time()
            else:
                if workout_active and time.time() - last_movement_time > 2:
                    rest_times.append((last_movement_time, time.time()))
                    last_movement_time = time.time()

        previous_landmarks = landmarks

        # Get key points for exercise calculations
        left_shoulder = [landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER.value].x,
                         landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER.value].y]
        right_shoulder = [landmarks[mp_pose.PoseLandmark.RIGHT_SHOULDER.value].x,
                          landmarks[mp_pose.PoseLandmark.RIGHT_SHOULDER.value].y]
        left_elbow = [landmarks[mp_pose.PoseLandmark.LEFT_ELBOW.value].x,
                      landmarks[mp_pose.PoseLandmark.LEFT_ELBOW.value].y]
        right_elbow = [landmarks[mp_pose.PoseLandmark.RIGHT_ELBOW.value].x,
                       landmarks[mp_pose.PoseLandmark.RIGHT_ELBOW.value].y]
        left_wrist = [landmarks[mp_pose.PoseLandmark.LEFT_WRIST.value].x,
                      landmarks[mp_pose.PoseLandmark.LEFT_WRIST.value].y]
        right_wrist = [landmarks[mp_pose.PoseLandmark.RIGHT_WRIST.value].x,
                       landmarks[mp_pose.PoseLandmark.RIGHT_WRIST.value].y]
        left_hip = [landmarks[mp_pose.PoseLandmark.LEFT_HIP.value].x,
                    landmarks[mp_pose.PoseLandmark.LEFT_HIP.value].y]
        right_hip = [landmarks[mp_pose.PoseLandmark.RIGHT_HIP.value].x,
                     landmarks[mp_pose.PoseLandmark.RIGHT_HIP.value].y]
        left_knee = [landmarks[mp_pose.PoseLandmark.LEFT_KNEE.value].x,
                     landmarks[mp_pose.PoseLandmark.LEFT_KNEE.value].y]
        right_knee = [landmarks[mp_pose.PoseLandmark.RIGHT_KNEE.value].x,
                      landmarks[mp_pose.PoseLandmark.RIGHT_KNEE.value].y]
        left_ankle = [landmarks[mp_pose.PoseLandmark.LEFT_ANKLE.value].x,
                      landmarks[mp_pose.PoseLandmark.LEFT_ANKLE.value].y]
        right_ankle = [landmarks[mp_pose.PoseLandmark.RIGHT_ANKLE.value].x,
                       landmarks[mp_pose.PoseLandmark.RIGHT_ANKLE.value].y]

        # Push-up detection logic
        left_elbow_angle = calculate_angle(left_shoulder, left_elbow, left_wrist)
        right_elbow_angle = calculate_angle(right_shoulder, right_elbow, right_wrist)
        if is_near_floor(left_wrist, left_elbow) and is_near_floor(right_wrist, right_elbow):
            if left_elbow_angle < 90 and right_elbow_angle < 90:
                pushup_stage = "down"
            if left_elbow_angle > 160 and right_elbow_angle > 160 and pushup_stage == "down":
                pushup_stage = "up"
                pushup_counter += 1

        # Squat detection logic
        left_knee_angle = calculate_angle(left_hip, left_knee, left_ankle)
        right_knee_angle = calculate_angle(right_hip, right_knee, right_ankle)
        if left_knee_angle < 90 and right_knee_angle < 90:
            squat_stage = "down"
        if left_knee_angle > 160 and right_knee_angle > 160 and squat_stage == "down":
            squat_stage = "up"
            squat_counter += 1

        # Pull-up detection logic
        left_shoulder_angle = calculate_angle(left_elbow, left_shoulder, left_hip)
        right_shoulder_angle = calculate_angle(right_elbow, right_shoulder, right_hip)
        if left_shoulder_angle < 90 and right_shoulder_angle < 90:
            pullup_stage = "up"
        if left_shoulder_angle > 160 and right_shoulder_angle > 160 and pullup_stage == "up":
            pullup_stage = "down"
            pullup_counter += 1

        # Draw pose landmarks
        mp_drawing.draw_landmarks(frame, results.pose_landmarks, mp_pose.POSE_CONNECTIONS)

    # Display workout stats
    cv2.rectangle(frame, (0, 0), (frame.shape[1], 120), (0, 0, 0), -1)
    cv2.putText(frame, f'Push-ups: {pushup_counter}', (10, 30),
                cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
    cv2.putText(frame, f'Pull-ups: {pullup_counter}', (200, 30),
                cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 0), 2)
    cv2.putText(frame, f'Squats: {squat_counter}', (400, 30),
                cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 0, 0), 2)

    # Show dynamic rest time during inactivity
    current_time = time.time()
    if current_time - last_movement_time > 2:  # Rest period ongoing
        rest_duration = int(current_time - last_movement_time)
        cv2.putText(frame, f'Rest Time: {rest_duration}s', (10, 80),
                    cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 255), 2)

    # End workout session after prolonged inactivity
    if workout_active and time.time() - last_movement_time > 10:
        end_time = time.time()
        workout_active = False

    cv2.imshow('Workout Tracker', frame)

    if cv2.waitKey(10) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()

# Print workout summary
if start_time:
    print(f"Workout started at: {time.strftime('%H:%M:%S', time.localtime(start_time))}")
if end_time:
    print(f"Workout ended at: {time.strftime('%H:%M:%S', time.localtime(end_time))}")
print(f"Total Push-ups: {pushup_counter}")
print(f"Total Pull-ups: {pullup_counter}")
print(f"Total Squats: {squat_counter}")
if rest_times:
    print("Rest times:")
    for start, end in rest_times:
        print(f"Rest from {time.strftime('%H:%M:%S', time.localtime(start))} to {time.strftime('%H:%M:%S', time.localtime(end))}")
```

```python
import cv2
import mediapipe as mp
import numpy as np
from tkinter import *
from PIL import Image, ImageTk
import time

# Mediapipe setup
mp_pose = mp.solutions.pose
mp_drawing = mp.solutions.drawing_utils
pose = mp_pose.Pose()

# Function to calculate angles
def calculate_angle(a, b, c):
    a = np.array(a)  # First point
    b = np.array(b)  # Second point (vertex)
    c = np.array(c)  # Third point

    radians = np.arctan2(c[1] - b[1], c[0] - b[0]) - np.arctan2(a[1] - b[1], a[0] - b[0])
    angle = np.abs(radians * 180.0 / np.pi)

    if angle > 180.0:
        angle = 360.0 - angle

    return angle

# Check if landmarks are visible
def are_landmarks_visible(landmarks):
    return all(landmark.visibility > 0.5 for landmark in landmarks)

# Check leg visibility for push-ups
def is_leg_visible(hip, knee, ankle):
    return hip[1] < knee[1] < ankle[1]

# App class
class WorkoutApp:
    def __init__(self, root):
        self.root = root
        self.root.title("Workout Counter")
       
        # Variables
        self.pushup_counter = 0
        self.pullup_counter = 0
        self.squat_counter = 0
        self.pushup_stage = None
        self.pullup_stage = None
        self.squat_stage = None
        self.cap = cv2.VideoCapture(0)  # Initialize video capture
        self.running = False
        self.rest_start_time = None
        self.is_resting = False

        # Create a label for video
        self.video_label = Label(root)
        self.video_label.pack()

        # Start video capture
        self.update_frame()

    def update_frame(self):
        ret, frame = self.cap.read()
        if not ret:
            print("Failed to capture video")
            return

        frame = cv2.flip(frame, 1)  # Flip the frame horizontally
        rgb_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        results = pose.process(rgb_frame)

        if results.pose_landmarks:
            landmarks = results.pose_landmarks.landmark

            if are_landmarks_visible(landmarks):
                # Reset rest timer
                self.is_resting = False
                self.rest_start_time = None

                # Key landmarks
                left_hip = [landmarks[mp_pose.PoseLandmark.LEFT_HIP.value].x,
                            landmarks[mp_pose.PoseLandmark.LEFT_HIP.value].y]
                left_knee = [landmarks[mp_pose.PoseLandmark.LEFT_KNEE.value].x,
                             landmarks[mp_pose.PoseLandmark.LEFT_KNEE.value].y]
                left_ankle = [landmarks[mp_pose.PoseLandmark.LEFT_ANKLE.value].x,
                              landmarks[mp_pose.PoseLandmark.LEFT_ANKLE.value].y]

                right_hip = [landmarks[mp_pose.PoseLandmark.RIGHT_HIP.value].x,
                             landmarks[mp_pose.PoseLandmark.RIGHT_HIP.value].y]
                right_knee = [landmarks[mp_pose.PoseLandmark.RIGHT_KNEE.value].x,
                              landmarks[mp_pose.PoseLandmark.RIGHT_KNEE.value].y]
                right_ankle = [landmarks[mp_pose.PoseLandmark.RIGHT_ANKLE.value].x,
                               landmarks[mp_pose.PoseLandmark.RIGHT_ANKLE.value].y]

                # Push-up detection
                left_shoulder = [landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER.value].x,
                                 landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER.value].y]
                left_elbow = [landmarks[mp_pose.PoseLandmark.LEFT_ELBOW.value].x,
                              landmarks[mp_pose.PoseLandmark.LEFT_ELBOW.value].y]
                left_wrist = [landmarks[mp_pose.PoseLandmark.LEFT_WRIST.value].x,
                              landmarks[mp_pose.PoseLandmark.LEFT_WRIST.value].y]

                right_shoulder = [landmarks[mp_pose.PoseLandmark.RIGHT_SHOULDER.value].x,
                                  landmarks[mp_pose.PoseLandmark.RIGHT_SHOULDER.value].y]
                right_elbow = [landmarks[mp_pose.PoseLandmark.RIGHT_ELBOW.value].x,
                               landmarks[mp_pose.PoseLandmark.RIGHT_ELBOW.value].y]
                right_wrist = [landmarks[mp_pose.PoseLandmark.RIGHT_WRIST.value].x,
                               landmarks[mp_pose.PoseLandmark.RIGHT_WRIST.value].y]

                # Check for push-ups
                if is_leg_visible(left_hip, left_knee, left_ankle) or is_leg_visible(right_hip, right_knee, right_ankle):
                    left_elbow_angle = calculate_angle(left_shoulder, left_elbow, left_wrist)
                    right_elbow_angle = calculate_angle(right_shoulder, right_elbow, right_wrist)

                    if left_elbow_angle < 90 and right_elbow_angle < 90:
                        self.pushup_stage = "down"
                    if left_elbow_angle > 160 and right_elbow_angle > 160 and self.pushup_stage == "down":
                        self.pushup_stage = "up"
                        self.pushup_counter += 1

                # Squat detection
                left_knee_angle = calculate_angle(left_hip, left_knee, left_ankle)
                right_knee_angle = calculate_angle(right_hip, right_knee, right_ankle)

                if left_knee_angle < 90 and right_knee_angle < 90:
                    self.squat_stage = "down"
                if left_knee_angle > 160 and right_knee_angle > 160 and self.squat_stage == "down":
                    self.squat_stage = "up"
                    self.squat_counter += 1

            else:
                # If full body is not detected, display warning
                cv2.putText(frame, 'Ensure full body is visible!', (10, 100),
                            cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2)
                # Start rest timer if no movement
                if not self.is_resting:
                    self.is_resting = True
                    self.rest_start_time = time.time()
                elif self.rest_start_time:
                    rest_time = int(time.time() - self.rest_start_time)
                    cv2.putText(frame, f'Rest Time: {rest_time}s', (10, 150),
                                cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 0), 2)

            # Display counters
            cv2.rectangle(frame, (0, 0), (frame.shape[1], 100), (0, 0, 0), -1)
            cv2.putText(frame, f'Push-ups: {self.pushup_counter}', (10, 30),
                        cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
            cv2.putText(frame, f'Pull-ups: {self.pullup_counter}', (200, 30),
                        cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 0), 2)
            cv2.putText(frame, f'Squats: {self.squat_counter}', (400, 30),
                        cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 255), 2)

        # Convert frame to ImageTk
        img = ImageTk.PhotoImage(image=Image.fromarray(rgb_frame))
        self.video_label.imgtk = img
        self.video_label.configure(image=img)

        self.root.after(10, self.update_frame)

    def on_closing(self):
        self.cap.release()
        self.root.destroy()

# Run the application
root = Tk()
app = WorkoutApp(root)
root.protocol("WM_DELETE_WINDOW", app.on_closing)  # Handle window close
root.mainloop()
```

```python
import time
import cv2
import numpy as np
import mediapipe as mp

# Initialize MediaPipe pose detector
mp_pose = mp.solutions.pose
mp_drawing = mp.solutions.drawing_utils
pose = mp_pose.Pose(static_image_mode=False, min_detection_confidence=0.5, min_tracking_confidence=0.5)

# Helper functions
def calculate_angle(a, b, c):
    radians = np.arctan2(c[1] - b[1], c[0] - b[0]) - np.arctan2(a[1] - b[1], a[0] - b[0])
    return np.abs(np.degrees(radians))

def is_near_floor(wrist, elbow, threshold=0.1):
    return abs(wrist[1] - elbow[1]) < threshold

def is_user_moving(current_landmarks, previous_landmarks):
    movement_threshold = 0.05
    for i in range(len(current_landmarks)):
        if abs(current_landmarks[i].x - previous_landmarks[i].x) > movement_threshold or \
           abs(current_landmarks[i].y - previous_landmarks[i].y) > movement_threshold:
            return True
    return False

def is_full_body_visible(landmarks):
    required_joints = [
        mp_pose.PoseLandmark.LEFT_SHOULDER.value,
        mp_pose.PoseLandmark.RIGHT_SHOULDER.value,
        mp_pose.PoseLandmark.LEFT_HIP.value,
        mp_pose.PoseLandmark.RIGHT_HIP.value,
        mp_pose.PoseLandmark.LEFT_KNEE.value,
        mp_pose.PoseLandmark.RIGHT_KNEE.value,
        mp_pose.PoseLandmark.LEFT_ANKLE.value,
        mp_pose.PoseLandmark.RIGHT_ANKLE.value
    ]
    for joint in required_joints:
        if landmarks[joint].visibility < 0.5:
            return False
    return True

# Initialize workout variables
pushup_counter, pullup_counter, squat_counter, bicep_curl_counter = 0, 0, 0, 0
pushup_stage, pullup_stage, squat_stage, curl_stage = None, None, None, None
start_time, end_time = None, None
last_movement_time = time.time()
rest_times = []
workout_active = False
previous_landmarks = None

# Start video capture
cap = cv2.VideoCapture(0)
cap.set(3, 1920)  # Full screen width
cap.set(4, 1080)  # Full screen height

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    frame = cv2.flip(frame, 1)
    rgb_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    results = pose.process(rgb_frame)

    if results.pose_landmarks:
        landmarks = results.pose_landmarks.landmark

        # Check if the full body is visible
        if not is_full_body_visible(landmarks):
            overlay = frame.copy()
            cv2.rectangle(overlay, (0, 0), (frame.shape[1], 100), (0, 0, 0), -1)
            alpha = 0.7
            cv2.addWeighted(overlay, alpha, frame, 1 - alpha, 0, frame)
            cv2.putText(frame, "Warning: Full body not visible", (10, 50),
                        cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2)
            cv2.imshow('Workout Tracker', frame)
            if cv2.waitKey(10) & 0xFF == ord('q'):
                break
            continue

        # Detect movement
        if previous_landmarks:
            if is_user_moving(landmarks, previous_landmarks):
                if not workout_active:
                    start_time = time.time()
                    workout_active = True
                last_movement_time = time.time()
            else:
                if workout_active and time.time() - last_movement_time > 2:
                    rest_times.append((last_movement_time, time.time()))
                    last_movement_time = time.time()

        previous_landmarks = landmarks

        # Get key points for exercise calculations
        left_shoulder = [landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER.value].x,
                         landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER.value].y]
        right_shoulder = [landmarks[mp_pose.PoseLandmark.RIGHT_SHOULDER.value].x,
                          landmarks[mp_pose.PoseLandmark.RIGHT_SHOULDER.value].y]
        left_elbow = [landmarks[mp_pose.PoseLandmark.LEFT_ELBOW.value].x,
                      landmarks[mp_pose.PoseLandmark.LEFT_ELBOW.value].y]
        right_elbow = [landmarks[mp_pose.PoseLandmark.RIGHT_ELBOW.value].x,
                       landmarks[mp_pose.PoseLandmark.RIGHT_ELBOW.value].y]
        left_wrist = [landmarks[mp_pose.PoseLandmark.LEFT_WRIST.value].x,
                      landmarks[mp_pose.PoseLandmark.LEFT_WRIST.value].y]
        right_wrist = [landmarks[mp_pose.PoseLandmark.RIGHT_WRIST.value].x,
                       landmarks[mp_pose.PoseLandmark.RIGHT_WRIST.value].y]
        left_hip = [landmarks[mp_pose.PoseLandmark.LEFT_HIP.value].x,
                    landmarks[mp_pose.PoseLandmark.LEFT_HIP.value].y]
        right_hip = [landmarks[mp_pose.PoseLandmark.RIGHT_HIP.value].x,
                     landmarks[mp_pose.PoseLandmark.RIGHT_HIP.value].y]
        left_knee = [landmarks[mp_pose.PoseLandmark.LEFT_KNEE.value].x,
                     landmarks[mp_pose.PoseLandmark.LEFT_KNEE.value].y]
        right_knee = [landmarks[mp_pose.PoseLandmark.RIGHT_KNEE.value].x,
                      landmarks[mp_pose.PoseLandmark.RIGHT_KNEE.value].y]
        left_ankle = [landmarks[mp_pose.PoseLandmark.LEFT_ANKLE.value].x,
                      landmarks[mp_pose.PoseLandmark.LEFT_ANKLE.value].y]
        right_ankle = [landmarks[mp_pose.PoseLandmark.RIGHT_ANKLE.value].x,
                       landmarks[mp_pose.PoseLandmark.RIGHT_ANKLE.value].y]

        # Push-up detection logic
        left_elbow_angle = calculate_angle(left_shoulder, left_elbow, left_wrist)
        right_elbow_angle = calculate_angle(right_shoulder, right_elbow, right_wrist)
        if is_near_floor(left_wrist, left_elbow) and is_near_floor(right_wrist, right_elbow):
            if left_elbow_angle < 90 and right_elbow_angle < 90:
                pushup_stage = "down"
            if left_elbow_angle > 160 and right_elbow_angle > 160 and pushup_stage == "down":
                pushup_stage = "up"
                pushup_counter += 1

        # Squat detection logic
        left_knee_angle = calculate_angle(left_hip, left_knee, left_ankle)
        right_knee_angle = calculate_angle(right_hip, right_knee, right_ankle)
        if left_knee_angle < 90 and right_knee_angle < 90:
            squat_stage = "down"
        if left_knee_angle > 160 and right_knee_angle > 160 and squat_stage == "down":
            squat_stage = "up"
            squat_counter += 1

        # Pull-up detection logic
        left_shoulder_angle = calculate_angle(left_elbow, left_shoulder, left_hip)
        right_shoulder_angle = calculate_angle(right_elbow, right_shoulder, right_hip)
        if left_shoulder_angle < 90 and right_shoulder_angle < 90:
            pullup_stage = "up"
        if left_shoulder_angle > 160 and right_shoulder_angle > 160 and pullup_stage == "up":
            pullup_stage = "down"
            pullup_counter += 1

        # Bicep curl detection logic
        left_bicep_angle = calculate_angle(left_shoulder, left_elbow, left_wrist)
        right_bicep_angle = calculate_angle(right_shoulder, right_elbow, right_wrist)
        if left_bicep_angle < 50 and right_bicep_angle < 50:
            curl_stage = "up"
        if left_bicep_angle > 160 and right_bicep_angle > 160 and curl_stage == "up":
            curl_stage = "down"
            bicep_curl_counter += 1

        # Draw pose landmarks
        mp_drawing.draw_landmarks(frame, results.pose_landmarks, mp_pose.POSE_CONNECTIONS)

    # Display workout stats
    overlay = frame.copy()
    cv2.rectangle(overlay, (0, 0), (frame.shape[1], 200), (0, 0, 0), -1)
    alpha = 0.7
    cv2.addWeighted(overlay, alpha, frame, 1 - alpha, 0, frame)
    cv2.putText(frame, f'Push-ups: {pushup_counter}', (10, 30),
                cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
    cv2.putText(frame, f'Pull-ups: {pullup_counter}', (10, 70),
                cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 0), 2)
    cv2.putText(frame, f'Squats: {squat_counter}', (10, 110),
                cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 0, 0), 2)
    cv2.putText(frame, f'Bicep Curls: {bicep_curl_counter}', (10, 150),
                cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 255), 2)

    # Show dynamic rest time during inactivity
    current_time = time.time()
    if current_time - last_movement_time > 2:  # Rest period ongoing
        rest_duration = int(current_time - last_movement_time)
        cv2.putText(frame, f'Rest Time: {rest_duration}s', (400, 70),
                    cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 255), 2)

    # End workout session after prolonged inactivity
    if workout_active and time.time() - last_movement_time > 10:
        end_time = time.time()
        workout_active = False

    cv2.imshow('Workout Tracker', frame)

    if cv2.waitKey(10) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()

# Print workout summary
if start_time:
    print(f"Workout started at: {time.strftime('%H:%M:%S', time.localtime(start_time))}")
if end_time:
    print(f"Workout ended at: {time.strftime('%H:%M:%S', time.localtime(end_time))}")
print(f"Total Push-ups: {pushup_counter}")
print(f"Total Pull-ups: {pullup_counter}")
print(f"Total Squats: {squat_counter}")
print(f"Total Bicep Curls: {bicep_curl_counter}")
if rest_times:
    print("Rest times:")
    for start, end in rest_times:
        print(f"Rest from {time.strftime('%H:%M:%S', time.localtime(start))} to {time.strftime('%H:%M:%S', time.localtime(end))}")
```

```python
import time
import cv2
import numpy as np
import mediapipe as mp

# Initialize MediaPipe pose detector
mp_pose = mp.solutions.pose
mp_drawing = mp.solutions.drawing_utils
pose = mp_pose.Pose(static_image_mode=False, min_detection_confidence=0.5, min_tracking_confidence=0.5)

# Helper functions
def calculate_angle(a, b, c):
    """Calculate angle between three points."""
    radians = np.arctan2(c[1] - b[1], c[0] - b[0]) - np.arctan2(a[1] - b[1], a[0] - b[0])
    angle = np.abs(np.degrees(radians))
    if angle > 180:
        angle = 360 - angle
    return angle

def is_near_floor(wrist, elbow, threshold=0.1):
    """Check if wrist and elbow are near the floor (for push-ups)."""
    return abs(wrist[1] - elbow[1]) < threshold

def is_user_moving(current_landmarks, previous_landmarks):
    """Check if the user is moving based on landmark positions."""
    movement_threshold = 0.05
    for i in range(len(current_landmarks)):
        if abs(current_landmarks[i].x - previous_landmarks[i].x) > movement_threshold or \
           abs(current_landmarks[i].y - previous_landmarks[i].y) > movement_threshold:
            return True
    return False

def is_full_body_visible(landmarks):
    """Check if the required body landmarks are visible."""
    required_joints = [
        mp_pose.PoseLandmark.LEFT_SHOULDER.value,
        mp_pose.PoseLandmark.RIGHT_SHOULDER.value,
        mp_pose.PoseLandmark.LEFT_HIP.value,
        mp_pose.PoseLandmark.RIGHT_HIP.value,
        mp_pose.PoseLandmark.LEFT_KNEE.value,
        mp_pose.PoseLandmark.RIGHT_KNEE.value,
        mp_pose.PoseLandmark.LEFT_ANKLE.value,
        mp_pose.PoseLandmark.RIGHT_ANKLE.value
    ]
    for joint in required_joints:
        if landmarks[joint].visibility < 0.5:
            return False
    return True

# Initialize workout variables
pushup_counter, pullup_counter, squat_counter = 0, 0, 0
right_counter, left_counter = 0, 0
pushup_stage, pullup_stage, squat_stage = None, None, None
right_stage, left_stage = None, None
start_time, end_time = None, None
last_movement_time = time.time()
rest_times = []
workout_active = False
previous_landmarks = None

# Start video capture
cap = cv2.VideoCapture(0)
cap.set(3, 1920)  # Full screen width
cap.set(4, 1080)  # Full screen height

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    frame = cv2.flip(frame, 1)
    rgb_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    results = pose.process(rgb_frame)

    if results.pose_landmarks:
        landmarks = results.pose_landmarks.landmark

        # Check if the full body is visible
        if not is_full_body_visible(landmarks):
            overlay = frame.copy()
            cv2.rectangle(overlay, (0, 0), (frame.shape[1], 100), (0, 0, 0), -1)
            alpha = 0.7
            cv2.addWeighted(overlay, alpha, frame, 1 - alpha, 0, frame)
            cv2.putText(frame, "Warning: Full body not visible", (10, 50),
                        cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2)
            cv2.imshow('Workout Tracker', frame)
            if cv2.waitKey(10) & 0xFF == ord('q'):
                break
            continue

        # Detect movement
        if previous_landmarks:
            if is_user_moving(landmarks, previous_landmarks):
                if not workout_active:
                    start_time = time.time()
                    workout_active = True
                last_movement_time = time.time()
            else:
                if workout_active and time.time() - last_movement_time > 2:
                    rest_times.append((last_movement_time, time.time()))
                    last_movement_time = time.time()

        previous_landmarks = landmarks

        # Get key points for exercise calculations
        left_shoulder = [landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER.value].x,
                         landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER.value].y]
        right_shoulder = [landmarks[mp_pose.PoseLandmark.RIGHT_SHOULDER.value].x,
                          landmarks[mp_pose.PoseLandmark.RIGHT_SHOULDER.value].y]
        left_elbow = [landmarks[mp_pose.PoseLandmark.LEFT_ELBOW.value].x,
                      landmarks[mp_pose.PoseLandmark.LEFT_ELBOW.value].y]
        right_elbow = [landmarks[mp_pose.PoseLandmark.RIGHT_ELBOW.value].x,
                       landmarks[mp_pose.PoseLandmark.RIGHT_ELBOW.value].y]
        left_wrist = [landmarks[mp_pose.PoseLandmark.LEFT_WRIST.value].x,
                      landmarks[mp_pose.PoseLandmark.LEFT_WRIST.value].y]
        right_wrist = [landmarks[mp_pose.PoseLandmark.RIGHT_WRIST.value].x,
                       landmarks[mp_pose.PoseLandmark.RIGHT_WRIST.value].y]
        left_hip = [landmarks[mp_pose.PoseLandmark.LEFT_HIP.value].x,
                    landmarks[mp_pose.PoseLandmark.LEFT_HIP.value].y]
        right_hip = [landmarks[mp_pose.PoseLandmark.RIGHT_HIP.value].x,
                     landmarks[mp_pose.PoseLandmark.RIGHT_HIP.value].y]
        left_knee = [landmarks[mp_pose.PoseLandmark.LEFT_KNEE.value].x,
                     landmarks[mp_pose.PoseLandmark.LEFT_KNEE.value].y]
        right_knee = [landmarks[mp_pose.PoseLandmark.RIGHT_KNEE.value].x,
                      landmarks[mp_pose.PoseLandmark.RIGHT_KNEE.value].y]
        left_ankle = [landmarks[mp_pose.PoseLandmark.LEFT_ANKLE.value].x,
                      landmarks[mp_pose.PoseLandmark.LEFT_ANKLE.value].y]
        right_ankle = [landmarks[mp_pose.PoseLandmark.RIGHT_ANKLE.value].x,
                       landmarks[mp_pose.PoseLandmark.RIGHT_ANKLE.value].y]

        # Push-up detection logic
        left_elbow_angle = calculate_angle(left_shoulder, left_elbow, left_wrist)
        right_elbow_angle = calculate_angle(right_shoulder, right_elbow, right_wrist)
        if is_near_floor(left_wrist, left_elbow) and is_near_floor(right_wrist, right_elbow):
            if left_elbow_angle < 90 and right_elbow_angle < 90:
                pushup_stage = "down"
            if left_elbow_angle > 160 and right_elbow_angle > 160 and pushup_stage == "down":
                pushup_stage = "up"
                pushup_counter += 1

        # Squat detection logic
        left_knee_angle = calculate_angle(left_hip, left_knee, left_ankle)
        right_knee_angle = calculate_angle(right_hip, right_knee, right_ankle)
        if left_knee_angle < 90 and right_knee_angle < 90:
            squat_stage = "down"
        if left_knee_angle > 160 and right_knee_angle > 160 and squat_stage == "down":
            squat_stage = "up"
            squat_counter += 1

        # Pull-up detection logic
        left_shoulder_angle = calculate_angle(left_elbow, left_shoulder, left_hip)
        right_shoulder_angle = calculate_angle(right_elbow, right_shoulder, right_hip)
        if left_shoulder_angle < 90 and right_shoulder_angle < 90:
            pullup_stage = "up"
        if left_shoulder_angle > 160 and right_shoulder_angle > 160 and pullup_stage == "up":
            pullup_stage = "down"
            pullup_counter += 1

        # Bicep curl detection logic
        right_bicep_angle = calculate_angle(right_shoulder, right_elbow, right_wrist)
        left_bicep_angle = calculate_angle(left_shoulder, left_elbow, left_wrist)

        # Right arm curl logic
        if right_bicep_angle > 160:
            right_stage = "down"
        if right_bicep_angle < 30 and right_stage == "down":
            right_stage = "up"
            right_counter += 1

        # Left arm curl logic
        if left_bicep_angle > 160:
            left_stage = "down"
        if left_bicep_angle < 30 and left_stage == "down":
            left_stage = "up"
            left_counter += 1

        # Draw pose landmarks (optional, can be commented out)
        # mp_drawing.draw_landmarks(frame, results.pose_landmarks, mp_pose.POSE_CONNECTIONS)

    # Display workout stats
    overlay = frame.copy()
    cv2.rectangle(overlay, (0, 0), (frame.shape[1], 200), (0, 0, 0), -1)
    alpha = 0.7
    cv2.addWeighted(overlay, alpha, frame, 1 - alpha, 0, frame)
    cv2.putText(frame, f'Push-ups: {pushup_counter}', (10, 30),
                cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
    cv2.putText(frame, f'Pull-ups: {pullup_counter}', (10, 70),
                cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 0), 2)
    cv2.putText(frame, f'Squats: {squat_counter}', (10, 110),
                cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 0, 0), 2)
    cv2.putText(frame, f'Right Curls: {right_counter}', (10, 150),
                cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 255), 2)
    cv2.putText(frame, f'Left Curls: {left_counter}', (10, 190),
                cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 255), 2)

    # Show dynamic rest time during inactivity
    current_time = time.time()
    if current_time - last_movement_time > 2:  # Rest period ongoing
        rest_duration = int(current_time - last_movement_time)
        cv2.putText(frame, f'Rest Time: {rest_duration}s', (400, 70),
                    cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 255), 2)

    # End workout session after prolonged inactivity
    if workout_active and time.time() - last_movement_time > 10:
        end_time = time.time()
        workout_active = False

    cv2.imshow('Workout Tracker', frame)

    if cv2.waitKey(10) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()

# Print workout summary
if start_time:
    print(f"Workout started at: {time.strftime('%H:%M:%S', time.localtime(start_time))}")
if end_time:
    print(f"Workout ended at: {time.strftime('%H:%M:%S', time.localtime(end_time))}")
print(f"Total Push-ups: {pushup_counter}")
print(f"Total Pull-ups: {pullup_counter}")
print(f"Total Squats: {squat_counter}")
print(f"Total Right Arm Curls: {right_counter}")
print(f"Total Left Arm Curls: {left_counter}")
if rest_times:
    print("Rest times:")
    for start, end in rest_times:
        print(f"Rest from {time.strftime('%H:%M:%S', time.localtime(start))} to {time.strftime('%H:%M:%S', time.localtime(end))}")
```

```python
import time
import cv2
import numpy as np
import mediapipe as mp
from reportlab.lib.pagesizes import letter
from reportlab.pdfgen import canvas
import sqlite3
from datetime import datetime
import os
from reportlab.lib import colors
from reportlab.graphics.shapes import Drawing
from reportlab.graphics.charts.linecharts import HorizontalLineChart
from reportlab.lib.styles import getSampleStyleSheet
from reportlab.platypus import SimpleDocTemplate, Paragraph, Spacer, Table, TableStyle

# Database setup
def setup_database():
    conn = sqlite3.connect('workout_history.db')
    c = conn.cursor()
    c.execute('''CREATE TABLE IF NOT EXISTS workouts
                 (date TEXT,
                  pushups INTEGER,
                  pullups INTEGER,
                  squats INTEGER,
                  right_curls INTEGER,
                  left_curls INTEGER,
                  duration INTEGER,
                  rest_time INTEGER)''')
    conn.commit()
    conn.close()

# Save workout data
def save_workout(pushups, pullups, squats, right_curls, left_curls, duration, rest_time):
    conn = sqlite3.connect('workout_history.db')
    c = conn.cursor()
    today = datetime.now().strftime('%Y-%m-%d')
   
    # Check if entry for today exists
    c.execute('SELECT * FROM workouts WHERE date = ?', (today,))
    existing = c.fetchone()
   
    if existing:
        # Update existing entry
        c.execute('''UPDATE workouts
                    SET pushups = pushups + ?,
                        pullups = pullups + ?,
                        squats = squats + ?,
                        right_curls = right_curls + ?,
                        left_curls = left_curls + ?,
                        duration = duration + ?,
                        rest_time = rest_time + ?
                    WHERE date = ?''',
                 (pushups, pullups, squats, right_curls, left_curls,
                  duration, rest_time, today))
    else:
        # Create new entry
        c.execute('''INSERT INTO workouts
                    VALUES (?, ?, ?, ?, ?, ?, ?, ?)''',
                 (today, pushups, pullups, squats, right_curls,
                  left_curls, duration, rest_time))
   
    conn.commit()
    conn.close()

def generate_progress_chart(data, exercise_type):
    """Generate a line chart showing progress over time"""
    drawing = Drawing(400, 200)
   
    chart = HorizontalLineChart()
    chart.x = 50
    chart.y = 50
    chart.width = 300
    chart.height = 150
   
    # Extract dates and values
    dates = [row[0] for row in data]
    values = [row[1] for row in data]
   
    chart.data = [values]
    chart.categoryAxis.categoryNames = dates
    chart.categoryAxis.labels.angle = 45
   
    drawing.add(chart)
    return drawing

def generate_pdf_report(start_date=None, end_date=None):
    """Generate a PDF report for a specific date range"""
    conn = sqlite3.connect('workout_history.db')
    c = conn.cursor()
   
    if start_date and end_date:
        c.execute('''SELECT * FROM workouts
                    WHERE date BETWEEN ? AND ?
                    ORDER BY date''', (start_date, end_date))
    else:
        c.execute('SELECT * FROM workouts ORDER BY date')
   
    data = c.fetchall()
    conn.close()
   
    if not data:
        print("No workout data found for the specified period")
        return
   
    # Create PDF
    report_time = datetime.now().strftime('%Y%m%d_%H%M%S')
    pdf_file = f"workout_report_{report_time}.pdf"
    doc = SimpleDocTemplate(pdf_file, pagesize=letter)
    styles = getSampleStyleSheet()
    elements = []
   
    # Title
    title = "Workout Progress Report"
    if start_date and end_date:
        title += f" ({start_date} to {end_date})"
    elements.append(Paragraph(title, styles['Title']))
    elements.append(Spacer(1, 20))
   
    # Summary table
    table_data = [['Date', 'Push-ups', 'Pull-ups', 'Squats', 'Right Curls', 'Left Curls',
                   'Duration (min)', 'Rest Time (min)']]
    for row in data:
        formatted_row = [row[0],  # date
                        str(row[1]),  # pushups
                        str(row[2]),  # pullups
                        str(row[3]),  # squats
                        str(row[4]),  # right_curls
                        str(row[5]),  # left_curls
                        str(round(row[6]/60, 1)),  # duration in minutes
                        str(round(row[7]/60, 1))]  # rest time in minutes
        table_data.append(formatted_row)
   
    # Calculate totals
    totals = ['Total']
    for i in range(1, 8):
        total = sum(float(row[i]) for row in table_data[1:])
        totals.append(str(round(total, 1)))
    table_data.append(totals)
   
    # Create and style the table
    table = Table(table_data)
    table.setStyle(TableStyle([
        ('BACKGROUND', (0, 0), (-1, 0), colors.grey),
        ('TEXTCOLOR', (0, 0), (-1, 0), colors.whitesmoke),
        ('ALIGN', (0, 0), (-1, -1), 'CENTER'),
        ('FONTNAME', (0, 0), (-1, 0), 'Helvetica-Bold'),
        ('FONTSIZE', (0, 0), (-1, 0), 14),
        ('BOTTOMPADDING', (0, 0), (-1, 0), 12),
        ('BACKGROUND', (0, -1), (-1, -1), colors.lightgrey),
        ('TEXTCOLOR', (0, -1), (-1, -1), colors.black),
        ('FONTNAME', (0, -1), (-1, -1), 'Helvetica-Bold'),
        ('GRID', (0, 0), (-1, -1), 1, colors.black)
    ]))
   
    elements.append(table)
    elements.append(Spacer(1, 20))
   
    # Add progress charts
    for i, exercise in enumerate(['Push-ups', 'Pull-ups', 'Squats', 'Right Curls', 'Left Curls']):
        elements.append(Paragraph(f"{exercise} Progress", styles['Heading2']))
        elements.append(Spacer(1, 10))
        chart_data = [(row[0], row[i+1]) for row in data]
        elements.append(generate_progress_chart(chart_data, exercise))
        elements.append(Spacer(1, 20))
   
    # Generate the PDF
    doc.build(elements)
    print(f"PDF report generated: {pdf_file}")

# Initialize MediaPipe pose detector
mp_pose = mp.solutions.pose
mp_drawing = mp.solutions.drawing_utils
pose = mp_pose.Pose(static_image_mode=False, min_detection_confidence=0.5, min_tracking_confidence=0.5)

# [Rest of the original helper functions remain the same]
# ... (keep all the helper functions from the original code)

# Modified main loop
def main():
    setup_database()
   
    # Initialize workout variables
    pushup_counter, pullup_counter, squat_counter = 0, 0, 0
    right_counter, left_counter = 0, 0
    pushup_stage, pullup_stage, squat_stage = None, None, None
    right_stage, left_stage = None, None
    start_time = time.time()
    last_movement_time = time.time()
    rest_times = []
    workout_active = False
    previous_landmarks = None

    # Start video capture
    cap = cv2.VideoCapture(0)
    cap.set(3, 1920)
    cap.set(4, 1080)

    try:
        while cap.isOpened():
            # [Original workout tracking code remains the same]
            # ... (keep all the original tracking code)
           
            if cv2.waitKey(10) & 0xFF == ord('q'):
                break
           
    finally:
        cap.release()
        cv2.destroyAllWindows()
       
        # Calculate workout stats
        end_time = time.time()
        total_duration = int(end_time - start_time)
        total_rest_time = sum(end - start for start, end in rest_times)
       
        # Save workout data
        save_workout(
            pushup_counter,
            pullup_counter,
            squat_counter,
            right_counter,
            left_counter,
            total_duration,
            int(total_rest_time)
        )
       
        # Generate today's report
        today = datetime.now().strftime('%Y-%m-%d')
        generate_pdf_report(today, today)

if __name__ == "__main__":
    main()
```

```python
import time
import cv2
import numpy as np
import mediapipe as mp
from reportlab.lib.pagesizes import letter
from reportlab.pdfgen import canvas
import sqlite3
from datetime import datetime
import os
from reportlab.lib import colors
from reportlab.graphics.shapes import Drawing
from reportlab.graphics.charts.linecharts import HorizontalLineChart
from reportlab.lib.styles import getSampleStyleSheet
from reportlab.platypus import SimpleDocTemplate, Paragraph, Spacer, Table, TableStyle

# Database setup
def setup_database():
    conn = sqlite3.connect('workout_history.db')
    c = conn.cursor()
    c.execute('''CREATE TABLE IF NOT EXISTS workouts
                 (date TEXT,
                  pushups INTEGER,
                  pullups INTEGER,
                  squats INTEGER,
                  right_curls INTEGER,
                  left_curls INTEGER,
                  duration INTEGER,
                  rest_time INTEGER)''')
    c.execute("CREATE INDEX IF NOT EXISTS idx_date ON workouts (date);")
    conn.commit()
    conn.close()

# Save workout data incrementally
def save_workout_incremental(pushups, pullups, squats, right_curls, left_curls, duration, rest_time):
    conn = sqlite3.connect('workout_history.db')
    c = conn.cursor()
    today = datetime.now().strftime('%Y-%m-%d')

    c.execute('SELECT * FROM workouts WHERE date = ?', (today,))
    existing = c.fetchone()

    if existing:
        c.execute('''UPDATE workouts
                      SET pushups = pushups + ?,
                          pullups = pullups + ?,
                          squats = squats + ?,
                          right_curls = right_curls + ?,
                          left_curls = left_curls + ?,
                          duration = duration + ?,
                          rest_time = rest_time + ?
                      WHERE date = ?''',
                  (pushups, pullups, squats, right_curls, left_curls,
                   duration, rest_time, today))
    else:
        c.execute('''INSERT INTO workouts
                      VALUES (?, ?, ?, ?, ?, ?, ?, ?)''',
                  (today, pushups, pullups, squats, right_curls,
                   left_curls, duration, rest_time))

    conn.commit()
    conn.close()

# Save workout data (Final Save)
def save_workout(pushups, pullups, squats, right_curls, left_curls, duration, rest_time):
    conn = sqlite3.connect('workout_history.db')
    c = conn.cursor()
    today = datetime.now().strftime('%Y-%m-%d')

    c.execute('SELECT * FROM workouts WHERE date = ?', (today,))
    existing = c.fetchone()

    if existing:
        c.execute('''UPDATE workouts
                      SET pushups = ?,
                          pullups = ?,
                          squats = ?,
                          right_curls = ?,
                          left_curls = ?,
                          duration = ?,
                          rest_time = ?
                      WHERE date = ?''',
                  (pushups, pullups, squats, right_curls, left_curls,
                   duration, rest_time, today))
    else:
        c.execute('''INSERT INTO workouts
                      VALUES (?, ?, ?, ?, ?, ?, ?, ?)''',
                  (today, pushups, pullups, squats, right_curls,
                   left_curls, duration, rest_time))

    conn.commit()
    conn.close()

def generate_progress_chart(data, exercise_type):
    drawing = Drawing(400, 200)
    chart = HorizontalLineChart()
    chart.x = 50
    chart.y = 50
    chart.width = 300
    chart.height = 150
    dates = [row[0] for row in data]
    values = [row[1] for row in data]
    chart.data = [values]
    chart.categoryAxis.categoryNames = dates
    chart.categoryAxis.labels.angle = 45
    drawing.add(chart)
    return drawing

def generate_pdf_report(start_date=None, end_date=None):
    conn = sqlite3.connect('workout_history.db')
    c = conn.cursor()

    if start_date and end_date:
        c.execute('''SELECT * FROM workouts
                     WHERE date BETWEEN ? AND ?
                     ORDER BY date''', (start_date, end_date))
    else:
        c.execute('SELECT * FROM workouts ORDER BY date')

    data = c.fetchall()
    conn.close()

    if not data:
        print("No workout data found for the specified period")
        return

    report_time = datetime.now().strftime('%Y%m%d_%H%M%S')
    pdf_file = f"workout_report_{report_time}.pdf"
    doc = SimpleDocTemplate(pdf_file, pagesize=letter)
    styles = getSampleStyleSheet()
    elements = []

    title = "Workout Progress Report"
    if start_date and end_date:
        title += f" ({start_date} to {end_date})"
    elements.append(Paragraph(title, styles['Title']))
    elements.append(Spacer(1, 20))

    table_data = [['Date', 'Push-ups', 'Pull-ups', 'Squats', 'Right Curls', 'Left Curls',
                   'Duration (min)', 'Rest Time (min)']]
    for row in data:
        formatted_row = [row[0], str(row[1]), str(row[2]), str(row[3]), str(row[4]), str(row[5]),
                         str(round(row[6] / 60, 1)), str(round(row[7] / 60, 1))]
        table_data.append(formatted_row)

    totals = ['Total']
    for i in range(1, 8):
        total = sum(float(row[i]) for row in table_data[1:])
        totals.append(str(round(total, 1)))
    table_data.append(totals)

    table = Table(table_data)
    table.setStyle(TableStyle([
        ('BACKGROUND', (0, 0), (-1, 0), colors.grey),
        ('TEXTCOLOR', (0, 0), (-1, 0), colors.whitesmoke),
        ('ALIGN', (0, 0), (-1, -1), 'CENTER'),
        ('FONTNAME', (0, 0), (-1, 0), 'Helvetica-Bold'),
        ('FONTSIZE', (0, 0), (-1, 0), 14),
        ('BOTTOMPADDING', (0, 0), (-1, 0), 12),
        ('BACKGROUND', (0, -1), (-1, -1), colors.lightgrey),
        ('TEXTCOLOR', (0, -1), (-1, -1), colors.black),
        ('FONTNAME', (0, -1), (-1, -1), 'Helvetica-Bold'),
        ('GRID', (0, 0), (-1, -1), 1, colors.black)
    ]))

    elements.append(table)
    elements.append(Spacer(1, 20))

    for i, exercise in enumerate(['Push-ups', 'Pull-ups', 'Squats', 'Right Curls', 'Left Curls']):
        elements.append(Paragraph(f"{exercise} Progress", styles['Heading2']))
        elements.append(Spacer(1, 10))
        chart_data = [(row[0], row[i + 1]) for row in data]
        elements.append(generate_progress_chart(chart_data, exercise))
        elements.append(Spacer(1, 20))

    doc.build(elements)
    print(f"PDF report generated: {pdf_file}")

# Initialize MediaPipe pose detector
mp_pose = mp.solutions.pose
mp_drawing = mp.solutions.drawing_utils
pose = mp_pose.Pose(static_image_mode=False, min_detection_confidence=0.5, min_tracking_confidence=0.5)

# Helper functions
def calculate_angle(a, b, c):
    a = np.array(a)
    b = np.array(b)
    c = np.array(c)
    radians = np.arctan2(c[1] - b[1], c[0] - b[0]) - np.arctan2(a[1] - b[1], a[0] - b[0])
    angle = np.abs(radians * 180.0 / np.pi)
    if angle > 180.0:
        angle = 360 - angle
    return angle

def get_coordinates(landmarks, point):
    try:
        return [landmarks[point].x, landmarks[point].y]
    except:
        return [None, None]

def main():
    setup_database()
    pushup_counter, pullup_counter, squat_counter = 0, 0, 0
    right_counter, left_counter = 0, 0
    pushup_stage, pullup_stage, squat_stage = None, None, None
    right_stage, left_stage = None, None
    start_time = time.time()
    last_movement_time = time.time()
    rest_times = []
    workout_active = False
    previous_landmarks = None
    frame_counter = 0

    cap = cv2.VideoCapture(0)
    cap.set(3, 640)
    cap.set(4, 480)

    try:
        while cap.isOpened():
            frame_counter += 1
            if frame_counter % 2 != 0:
                continue

            ret, frame = cap.read()
            if not ret:
                break

            image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
            image.flags.writeable = False
            results = pose.process(image)
            image.flags.writeable = True
            image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)

            if results.pose_landmarks:
                landmarks = results.pose_landmarks.landmark
                mp_drawing.draw_landmarks(image, results.pose_landmarks, mp_pose.POSE_CONNECTIONS)

                # Push-up detection
                left_shoulder = get_coordinates(landmarks, mp_pose.PoseLandmark.LEFT_SHOULDER.value)
                left_elbow = get_coordinates(landmarks, mp_pose.PoseLandmark.LEFT_ELBOW.value)
                left_wrist = get_coordinates(landmarks, mp_pose.PoseLandmark.LEFT_WRIST.value)

                if all(left_shoulder) and all(left_elbow) and all(left_wrist):
                    pushup_angle = calculate_angle(left_shoulder, left_elbow, left_wrist)
                    if pushup_angle > 160:
                        pushup_stage = "up"
                    if pushup_angle < 90 and pushup_stage == 'up':
                        pushup_stage = "down"
                        pushup_counter += 1
                        last_movement_time = time.time()
                        save_workout_incremental(1, 0, 0, 0, 0, 0, 0) #incremental save

                # Pull-up detection
                right_shoulder = get_coordinates(landmarks, mp_pose.PoseLandmark.RIGHT_SHOULDER.value)
                right_elbow = get_coordinates(landmarks, mp_pose.PoseLandmark.RIGHT_ELBOW.value)
                right_wrist = get_coordinates(landmarks, mp_pose.PoseLandmark.RIGHT_WRIST.value)

                if all(right_shoulder) and all(right_elbow) and all(right_wrist):
                    pullup_angle = calculate_angle(right_shoulder, right_elbow, right_wrist)
                    if pullup_angle > 160:
                        pullup_stage = "down"
                    if pullup_angle < 90 and pullup_stage == 'down':
                        pullup_stage = "up"
                        pullup_counter += 1
                        last_movement_time = time.time()
                        save_workout_incremental(0, 1, 0, 0, 0, 0, 0) #incremental save

                # Squat detection
                left_hip = get_coordinates(landmarks, mp_pose.PoseLandmark.LEFT_HIP.value)
                left_knee = get_coordinates(landmarks, mp_pose.PoseLandmark.LEFT_KNEE.value)
                left_ankle = get_coordinates(landmarks, mp_pose.PoseLandmark.LEFT_ANKLE.value)

                if all(left_hip) and all(left_knee) and all(left_ankle):
                    squat_angle = calculate_angle(left_hip, left_knee, left_ankle)
                    if squat_angle > 160:
                        squat_stage = "up"
                    if squat_angle < 90 and squat_stage == 'up':
                        squat_stage = "down"
                        squat_counter += 1
                        last_movement_time = time.time()
                        save_workout_incremental(0, 0, 1, 0, 0, 0, 0) #incremental save

                # Bicep curl detection (right arm)
                right_shoulder = get_coordinates(landmarks, mp_pose.PoseLandmark.RIGHT_SHOULDER.value)
                right_elbow = get_coordinates(landmarks, mp_pose.PoseLandmark.RIGHT_ELBOW.value)
                right_wrist = get_coordinates(landmarks, mp_pose.PoseLandmark.RIGHT_WRIST.value)

                if all(right_shoulder) and all(right_elbow) and all(right_wrist):
                    right_curl_angle = calculate_angle(right_shoulder, right_elbow, right_wrist)
                    if right_curl_angle > 160:
                        right_stage = "down"
                    if right_curl_angle < 50 and right_stage == 'down':
                        right_stage = "up"
                        right_counter += 1
                        last_movement_time = time.time()
                        save_workout_incremental(0, 0, 0, 1, 0, 0, 0) #incremental save

                # Bicep curl detection (left arm)
                left_shoulder = get_coordinates(landmarks, mp_pose.PoseLandmark.LEFT_SHOULDER.value)
                left_elbow = get_coordinates(landmarks, mp_pose.PoseLandmark.LEFT_ELBOW.value)
                left_wrist = get_coordinates(landmarks, mp_pose.PoseLandmark.LEFT_WRIST.value)

                if all(left_shoulder) and all(left_elbow) and all(left_wrist):
                    left_curl_angle = calculate_angle(left_shoulder, left_elbow, left_wrist)
                    if left_curl_angle > 160:
                        left_stage = "down"
                    if left_curl_angle < 50 and left_stage == 'down':
                        left_stage = "up"
                        left_counter += 1
                        last_movement_time = time.time()
                        save_workout_incremental(0, 0, 0, 0, 1, 0, 0) #incremental save

                # Rest time detection
                if previous_landmarks:
                    movement = False
                    for i in range(len(landmarks)):
                        if abs(landmarks[i].x - previous_landmarks[i].x) > 0.01 or abs(landmarks[i].y - previous_landmarks[i].y) > 0.01:
                            movement = True
                            break
                    if movement:
                        last_movement_time = time.time()
                    else:
                        if time.time() - last_movement_time > 10 and workout_active:
                            if not rest_times or (rest_times and rest_times[-1][1] is not None):
                                rest_times.append([time.time() - 10, None])
                previous_landmarks = landmarks

                workout_active = True

            cv2.imshow('Workout Tracking', image)

            if cv2.waitKey(10) & 0xFF == ord('q'):
                break

    finally:
        cap.release()
        cv2.destroyAllWindows()

        end_time = time.time()
        total_duration = int(end_time - start_time)
        total_rest_time = 0
        for start, end in rest_times:
            if end is None:
                total_rest_time += end_time - start
            else:
                total_rest_time += end - start

        save_workout(
            pushup_counter,
            pullup_counter,
            squat_counter,
            right_counter,
            left_counter,
            total_duration,
            int(total_rest_time)
        )

        today = datetime.now().strftime('%Y-%m-%d')
        generate_pdf_report(today, today)

if __name__ == "__main__":
    main()
```

```python
import time
import cv2
import numpy as np
import mediapipe as mp
from reportlab.lib.pagesizes import letter
from reportlab.pdfgen import canvas
import sqlite3
from datetime import datetime
import os
from reportlab.lib import colors
from reportlab.graphics.shapes import Drawing
from reportlab.graphics.charts.linecharts import HorizontalLineChart
from reportlab.lib.styles import getSampleStyleSheet
from reportlab.platypus import SimpleDocTemplate, Paragraph, Spacer, Table, TableStyle
from reportlab.graphics.charts.legends import Legend

# Initialize MediaPipe pose detector
mp_pose = mp.solutions.pose
mp_drawing = mp.solutions.drawing_utils
pose = mp_pose.Pose(
    static_image_mode=False,
    model_complexity=1,
    smooth_landmarks=True,
    min_detection_confidence=0.5,
    min_tracking_confidence=0.5
)

def setup_database():
    """Setup SQLite database with efficient indexing"""
    conn = sqlite3.connect('workout_history.db')
    c = conn.cursor()
    c.execute('''CREATE TABLE IF NOT EXISTS workouts
                 (date TEXT PRIMARY KEY,
                  pushups INTEGER,
                  pullups INTEGER,
                  squats INTEGER,
                  right_curls INTEGER,
                  left_curls INTEGER,
                  duration INTEGER,
                  rest_time INTEGER)''')
    c.execute('CREATE INDEX IF NOT EXISTS idx_date ON workouts(date)')
    conn.commit()
    conn.close()

def calculate_angle(a, b, c):
    """Calculate angle between three points"""
    a = np.array(a)
    b = np.array(b)
    c = np.array(c)
   
    radians = np.arctan2(c[1] - b[1], c[0] - b[0]) - np.arctan2(a[1] - b[1], a[0] - b[0])
    angle = np.abs(np.degrees(radians))
   
    if angle > 180:
        angle = 360 - angle
       
    return angle

def is_on_floor(landmarks, mp_pose):
    """Check if the person is in push-up position (on floor)"""
    # Get relevant landmarks
    left_wrist = landmarks[mp_pose.PoseLandmark.LEFT_WRIST.value].y
    right_wrist = landmarks[mp_pose.PoseLandmark.RIGHT_WRIST.value].y
    left_ankle = landmarks[mp_pose.PoseLandmark.LEFT_ANKLE.value].y
    right_ankle = landmarks[mp_pose.PoseLandmark.RIGHT_ANKLE.value].y
    left_hip = landmarks[mp_pose.PoseLandmark.LEFT_HIP.value].y
    right_hip = landmarks[mp_pose.PoseLandmark.RIGHT_HIP.value].y
   
    # Check if body is horizontal and hands are on ground
    body_horizontal = abs(left_hip - right_hip) < 0.1
    hands_down = (left_wrist > 0.7 and right_wrist > 0.7)
   
    return body_horizontal and hands_down

def is_standing(landmarks, mp_pose):
    """Check if person is in standing position"""
    hip_y = (landmarks[mp_pose.PoseLandmark.LEFT_HIP.value].y +
             landmarks[mp_pose.PoseLandmark.RIGHT_HIP.value].y) / 2
    ankle_y = (landmarks[mp_pose.PoseLandmark.LEFT_ANKLE.value].y +
               landmarks[mp_pose.PoseLandmark.RIGHT_ANKLE.value].y) / 2
    shoulder_y = (landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER.value].y +
                 landmarks[mp_pose.PoseLandmark.RIGHT_SHOULDER.value].y) / 2
   
    vertical_alignment = abs(hip_y - ankle_y) > 0.3
    upright_posture = shoulder_y < hip_y
   
    return vertical_alignment and upright_posture

def is_full_body_visible(landmarks):
    """Check if the required body landmarks are visible"""
    required_joints = [
        mp_pose.PoseLandmark.LEFT_SHOULDER.value,
        mp_pose.PoseLandmark.RIGHT_SHOULDER.value,
        mp_pose.PoseLandmark.LEFT_HIP.value,
        mp_pose.PoseLandmark.RIGHT_HIP.value,
        mp_pose.PoseLandmark.LEFT_KNEE.value,
        mp_pose.PoseLandmark.RIGHT_KNEE.value,
        mp_pose.PoseLandmark.LEFT_ANKLE.value,
        mp_pose.PoseLandmark.RIGHT_ANKLE.value
    ]
   
    for joint in required_joints:
        if landmarks[joint].visibility < 0.5:
            return False
    return True

def generate_progress_chart(data, exercise_type):
    """Generate an improved line chart showing progress over time"""
    drawing = Drawing(500, 250)
   
    lc = HorizontalLineChart()
    lc.x = 50
    lc.y = 50
    lc.width = 400
    lc.height = 150
   
    # Extract dates and values
    dates = [row[0] for row in data]
    values = [row[1] for row in data]
   
    # Configure chart
    lc.data = [values]
    lc.categoryAxis.categoryNames = dates
    lc.categoryAxis.labels.angle = 45
    lc.categoryAxis.labels.fontSize = 8
   
    # Style improvements
    lc.lines[0].strokeWidth = 2
    lc.lines[0].strokeColor = colors.blue
    lc.valueAxis.gridStrokeColor = colors.lightgrey
    lc.valueAxis.gridStrokeWidth = 0.5
    lc.valueAxis.labelTextFormat = '%d'
   
    # Add markers
    lc.lines.symbol = 'CircleY'
    lc.lines.symbol_size = 5
   
    # Add legend
    legend = Legend()
    legend.x = 50
    legend.y = 220
    legend.alignment = 'right'
    legend.columnMaximum = 1
    legend.colorNamePairs = [(colors.blue, exercise_type)]
    legend.fontSize = 8
   
    drawing.add(lc)
    drawing.add(legend)
   
    return drawing

def generate_pdf_report(start_date=None, end_date=None):
    """Generate PDF report with improved graphs"""
    conn = sqlite3.connect('workout_history.db')
    c = conn.cursor()
   
    if start_date and end_date:
        c.execute('''SELECT * FROM workouts
                    WHERE date BETWEEN ? AND ?
                    ORDER BY date''', (start_date, end_date))
    else:
        c.execute('SELECT * FROM workouts ORDER BY date')
   
    data = c.fetchall()
    conn.close()
   
    if not data:
        print("No workout data found for the specified period")
        return
   
    report_time = datetime.now().strftime('%Y%m%d_%H%M%S')
    pdf_file = f"workout_report_{report_time}.pdf"
    doc = SimpleDocTemplate(pdf_file, pagesize=letter,
                          rightMargin=50, leftMargin=50,
                          topMargin=50, bottomMargin=50)
   
    elements = []
    styles = getSampleStyleSheet()
   
    # Title
    title = "Workout Progress Report"
    if start_date and end_date:
        title += f"\n{start_date} to {end_date}"
    elements.append(Paragraph(title, styles['Title']))
    elements.append(Spacer(1, 30))
   
    # Summary table
    table_data = [['Date', 'Push-ups', 'Pull-ups', 'Squats', 'Right Curls', 'Left Curls',
                   'Duration (min)', 'Rest Time (min)']]
   
    for row in data:
        formatted_row = [
            row[0],
            str(row[1]),
            str(row[2]),
            str(row[3]),
            str(row[4]),
            str(row[5]),
            str(round(row[6]/60, 1)),
            str(round(row[7]/60, 1))
        ]
        table_data.append(formatted_row)
   
    table = Table(table_data)
    table.setStyle(TableStyle([
        ('BACKGROUND', (0, 0), (-1, 0), colors.grey),
        ('TEXTCOLOR', (0, 0), (-1, 0), colors.whitesmoke),
        ('ALIGN', (0, 0), (-1, -1), 'CENTER'),
        ('FONTNAME', (0, 0), (-1, 0), 'Helvetica-Bold'),
        ('FONTSIZE', (0, 0), (-1, 0), 14),
        ('BOTTOMPADDING', (0, 0), (-1, 0), 12),
        ('GRID', (0, 0), (-1, -1), 1, colors.black)
    ]))
   
    elements.append(table)
    elements.append(Spacer(1, 30))
   
    # Add charts
    exercise_names = ['Push-ups', 'Pull-ups', 'Squats', 'Right Curls', 'Left Curls']
    for i, exercise in enumerate(exercise_names):
        elements.append(Paragraph(f"{exercise} Progress", styles['Heading2']))
        elements.append(Spacer(1, 10))
        chart_data = [(row[0], row[i+1]) for row in data]
        elements.append(generate_progress_chart(chart_data, exercise))
        elements.append(Spacer(1, 30))
   
    doc.build(elements)
    print(f"PDF report generated: {pdf_file}")

def save_workout(pushups, pullups, squats, right_curls, left_curls, duration, rest_time):
    """Save workout data efficiently"""
    conn = sqlite3.connect('workout_history.db')
    c = conn.cursor()
    today = datetime.now().strftime('%Y-%m-%d')
   
    c.execute('''INSERT INTO workouts VALUES (?, ?, ?, ?, ?, ?, ?, ?)
                ON CONFLICT(date) DO UPDATE SET
                pushups = pushups + ?,
                pullups = pullups + ?,
                squats = squats + ?,
                right_curls = right_curls + ?,
                left_curls = left_curls + ?,
                duration = duration + ?,
                rest_time = rest_time + ?''',
             (today, pushups, pullups, squats, right_curls, left_curls, duration, rest_time,
              pushups, pullups, squats, right_curls, left_curls, duration, rest_time))
   
    conn.commit()
    conn.close()

def main():
    setup_database()
   
    # Initialize counters and states
    pushup_counter = pullup_counter = squat_counter = right_counter = left_counter = 0
    pushup_stage = pullup_stage = squat_stage = right_stage = left_stage = None
   
    # Initialize timing variables
    start_time = time.time()
    last_movement_time = time.time()
    rest_times = []
    workout_active = False
   
    # Video capture settings
    cap = cv2.VideoCapture(0)
    cap.set(cv2.CAP_PROP_FRAME_WIDTH, 640)
    cap.set(cv2.CAP_PROP_FRAME_HEIGHT, 480)
    cap.set(cv2.CAP_PROP_FPS, 30)
   
    frame_count = 0
    process_every_n_frames = 2
   
    try:
        while cap.isOpened():
            ret, frame = cap.read()
            if not ret:
                break
               
            frame_count += 1
            if frame_count % process_every_n_frames != 0:
                continue
               
            frame = cv2.flip(frame, 1)
            rgb_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
            rgb_frame.flags.writeable = False
            results = pose.process(rgb_frame)
            rgb_frame.flags.writeable = True
           
            if results.pose_landmarks:
                landmarks = results.pose_landmarks.landmark
               
                if not is_full_body_visible(landmarks):
                    cv2.putText(frame, "Please step back - Full body not visible",
                              (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2)
                    cv2.imshow('Workout Tracker', frame)
                    continue
               
                # Get key points
                left_shoulder = [landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER.value].x,
                               landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER.value].y]
                right_shoulder = [landmarks[mp_pose.PoseLandmark.RIGHT_SHOULDER.value].x,
                                landmarks[mp_pose.PoseLandmark.RIGHT_SHOULDER.value].y]
                left_elbow = [landmarks[mp_pose.PoseLandmark.LEFT_ELBOW.value].x,
                            landmarks[mp_pose.PoseLandmark.LEFT_ELBOW.value].y]
                right_elbow = [landmarks[mp_pose.PoseLandmark.RIGHT_ELBOW.value].x,
                             landmarks[mp_pose.PoseLandmark.RIGHT_ELBOW.value].y]
                left_wrist = [landmarks[mp_pose.PoseLandmark.LEFT_WRIST.value].x,
                            landmarks[mp_pose.PoseLandmark.LEFT_WRIST.value].y]
                right_wrist = [landmarks[mp_pose.PoseLandmark.RIGHT_WRIST.value].x,
                             landmarks[mp_pose.PoseLandmark.RIGHT_WRIST.value].y]
               
                # Calculate angles
                left_elbow_angle = calculate_angle(left_shoulder, left_elbow, left_wrist)
                right_elbow_angle = calculate_angle(right_shoulder, right_elbow, right_wrist)
               
                # Push-up detection (only when on floor)
                if is_on_floor(landmarks, mp_pose):
                    if left_elbow_angle < 90 and right_elbow_angle < 90:
                        pushup_stage = "down"
                    elif left_elbow_angle > 160 and right_elbow_angle > 160 and pushup_stage == "down":
                        pushup_stage = "up"
                        pushup_counter += 1
                        cv2.putText(frame, "Push-up counted!", (10, 60),
                                  cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
               
                # Bicep curl detection (only when standing)
                elif is_standing(landmarks, mp_pose):
                    # Right arm curl
                    if right_elbow_angle > 160:
                        right_stage = "down"
                    elif right_elbow_angle < 30 and right_stage == "down":
                        right_stage = "up"
                        right_counter += 1
# Left arm curl
                    if left_elbow_angle > 160:
                        left_stage = "down"
                    elif left_elbow_angle < 30 and left_stage == "down":
                        left_stage = "up"
                        left_counter += 1
                       
                # Draw pose landmarks
                mp_drawing.draw_landmarks(
                    frame,
                    results.pose_landmarks,
                    mp_pose.POSE_CONNECTIONS,
                    mp_drawing.DrawingSpec(color=(245,117,66), thickness=2, circle_radius=2),
                    mp_drawing.DrawingSpec(color=(245,66,230), thickness=2, circle_radius=2)
                )
               
                # Display exercise status
                # Create semi-transparent overlay for stats
                overlay = frame.copy()
                cv2.rectangle(overlay, (0, 0), (350, 200), (0, 0, 0), -1)
                alpha = 0.7
                frame = cv2.addWeighted(overlay, alpha, frame, 1 - alpha, 0)

                # Display counters
                cv2.putText(frame, f'Push-ups: {pushup_counter}',
                          (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
                cv2.putText(frame, f'Pull-ups: {pullup_counter}',
                          (10, 70), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 0), 2)
                cv2.putText(frame, f'Squats: {squat_counter}',
                          (10, 110), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 0, 0), 2)
                cv2.putText(frame, f'Right Curls: {right_counter}',
                          (10, 150), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 255), 2)
                cv2.putText(frame, f'Left Curls: {left_counter}',
                          (10, 190), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 255), 2)

                # Display current position status
                position_status = ""
                if is_on_floor(landmarks, mp_pose):
                    position_status = "Push-up Position"
                elif is_standing(landmarks, mp_pose):
                    position_status = "Standing Position"
               
                if position_status:
                    cv2.putText(frame, f'Position: {position_status}',
                              (10, 230), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 255), 2)

            # Display the frame
            cv2.imshow('Workout Tracker', frame)

            # Break the loop if 'q' is pressed
            if cv2.waitKey(1) & 0xFF == ord('q'):
                break

    finally:
        # Calculate final workout stats
        end_time = time.time()
        total_duration = int(end_time - start_time)
        total_rest_time = sum(end - start for start, end in rest_times)

        # Save workout data
        save_workout(
            pushup_counter,
            pullup_counter,
            squat_counter,
            right_counter,
            left_counter,
            total_duration,
            int(total_rest_time)
        )

        # Generate report for today
        today = datetime.now().strftime('%Y-%m-%d')
        generate_pdf_report(today, today)

        # Cleanup
        cap.release()
        cv2.destroyAllWindows()

if __name__ == "__main__":
    main()
```
